**Project Objective
To build a machine learning model that can classify whether a tumor is Malignant (M) or Benign (B) based on its characteristics**

*Step-by-Step Implementation*

1. Import Required Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

2. Load the Dataset

```python
df = pd.read_csv('/content/cancer_data/Cancer_Data.csv')
df.head()
```

3. Drop Unnecessary Columns


*If id or unnamed index column is present*

```python
df.drop(['id'], axis=1, inplace=True)  # If 'id' is not useful
```

4. Encode the Target Variable

```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])  # M -> 1, B -> 0
```

5. Split Features and Target

```python
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

6. Scale the Data

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

7. Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

8. Train a Classifier

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

9. Evaluate the Model

```python
y_pred = model.predict(X_test)

print("Classification Report:\n", classification_report(y_test, y_pred))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='coolwarm')
plt.title("Confusion Matrix")
plt.show()
```

Feature Importance

```python
feature_importances = pd.Series(model.feature_importances_, index=X.columns)
feature_importances.nlargest(10).plot(kind='barh')
plt.title("Top 10 Important Features")
plt.show()
```
